```python
severity={"Minor_Damage_And_Injuries":0,"Significant_Damage_And_Serious_Injuries":1,"Significant_Damage_And_Fatalities":2,"Highly_Fatal_And_Damaging":3}
train.Severity=train['Severity'].map(severity).astype(int)
```

```python
train['Safety_Score']=(train['Safety_Score'])**0.5
train['Adverse_Weather_Metric']=(train['Adverse_Weather_Metric'])**2
train['Cabin_Temperature']=(train['Cabin_Temperature'])**2
train['Control_Metric']=(train['Control_Metric'])**2
train['Days_Since_Inspection']=(train['Days_Since_Inspection'])**0.5
train['Total_Safety_Complaints']=(train['Total_Safety_Complaints'])**0.5
train['Turbulence_In_gforces']=(train['Turbulence_In_gforces'])**0.5
train['Max_Elevation']=(train['Max_Elevation'])**0.5
```

```python
test['Safety_Score']=(test['Safety_Score'])**0.5
test['Adverse_Weather_Metric']=(test['Adverse_Weather_Metric'])**2
test['Cabin_Temperature']=(test['Cabin_Temperature'])**2
test['Control_Metric']=(test['Control_Metric'])**2
test['Days_Since_Inspection']=(test['Days_Since_Inspection'])**0.5
test['Total_Safety_Complaints']=(test['Total_Safety_Complaints'])**0.5
test['Turbulence_In_gforces']=(test['Turbulence_In_gforces'])**0.5
test['Max_Elevation']=(test['Max_Elevation'])**0.5
```

```python
x=pd.DataFrame(train.iloc[:,1:-1])
               
y=pd.DataFrame(train.iloc[:,0])
```

```python
selected=SelectKBest(f_classif,k=10)
selected.fit(x,y)
sel_col=x.columns[selected.get_support()]
selected.get_support
```

```python
sel_col
x_sel=x[sel_col].drop(columns=['Accident_Type_Code'])
```

```python
x_tr,x_ts,y_tr,y_ts=train_test_split(x_sel,y,test_size=0.2,random_state=55,stratify=y)
```

```python
rf=RandomForestClassifier(n_estimators=1800,random_state=555,bootstrap=True)
rf.fit(x_tr,y_tr)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_tr,y_tr)
```

```python
ad=AdaBoostClassifier(base_estimator=rf,n_estimators=1500,random_state=555)
ad.fit(x_tr,y_tr)
```

```python
f1_score(y_ts,rf.predict(x_ts),average='weighted'),f1_score(y_ts,dt.predict(x_ts),average='weighted'),f1_score(y_ts,ad.predict(x_ts),average='weighted')
```

```python
op1=pd.DataFrame(rf.predict(test[sel_col].drop(columns=['Accident_Type_Code'])))
op2=pd.DataFrame(dt.predict(test[sel_col].drop(columns=['Accident_Type_Code'])))
op3=pd.DataFrame(ad.predict(test[sel_col].drop(columns=['Accident_Type_Code'])))
op4=pd.concat([op1,op2,op3],axis=1)
#op4['m']=op4.mode(axis
#op4.mode(axis=1).as
op4
```

```python
#op=pd.DataFrame(ad.predict(test[sel_col].drop(columns=['Accident_Type_Code'])))
op=op3
op.columns=['Severity']
inverse={0:"Minor_Damage_And_Injuries",1:"Significant_Damage_And_Serious_Injuries",2:"Significant_Damage_And_Fatalities",3:"Highly_Fatal_And_Damaging"}
op.Severity=op['Severity'].map(inverse).astype(str)
op.head()
```

```python
sev=pd.DataFrame()
sev['Accident_ID']=test['Accident_ID']
sev['Severity']=op
sev.to_csv("sample_submission.csv",index=False)
```
